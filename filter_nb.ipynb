```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
root = os.path.dirname(os.path.abspath("filter_nb.ipynb"))
root
```

```python
df = pd.read_csv(os.path.join(root, "pokemon_cards_csv_scape.csv"))
df
```

```python
df_removed_grade_more_than_10 = df[df['grade'] <= 10]
```

```python
sns.set_theme()
plt.figure(figsize=(10,6))
sns.scatterplot(x='grade', y='price', data=df_removed_grade_more_than_10, marker="o", color="dodgerblue")
plt.title("Grade vs Price for Pokemon Cards")
plt.show()
```

```python
df['grading_company'].value_counts(dropna=False).plot(kind='bar', color='darkorange', figsize=(10,6), title="Distribution of Grading Companies", xlabel="Grading Company", ylabel="Number of Cards")
plt.show()
```
